```python
import pandas as pd
iris = pd.read_csv('/content/Iris (3).csv', names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species'], skiprows=1)
iris.head()
iris.tail()
iris.describe()
```

```python
import matplotlib.pyplot as plt
x=iris['petal_length']
y=iris['petal_width']
plt.scatter(x,y)
plt.title("Scatter plot Petal length vs width")
plt.xlabel("petal length")
plt.ylabel("Petal width")

```

```python
import matplotlib.pyplot as plt
fig, x = plt.subplots()
# scatter the sepal_length against the sepal_width
x.scatter(iris['sepal_length'], iris['sepal_width'])
# set a title and labels
x.set_title('Iris Dataset')
x.set_xlabel('sepal_length')
x.set_ylabel('sepal_width')
```

```python
import matplotlib.pyplot as plt
plt.hist(iris['petal_length'], bins=10, color='pink', edgecolor='red')
plt.title("Distribution of Petal Length")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
